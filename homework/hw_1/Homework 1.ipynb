### Часть I

**Задача 1:**

Произошло убийство. На месте убийства найдена кровь, которая явно (будем считать, что с вероятностью 1) принадлежит настоящему убийце. Кровь принадлежит редкой группе, которая присутствует всего лишь у 1% населения. И так получилось, что у подсудимого как раз эта редкая группа крови!
<br>
Первым выходит прокурор и говорит: “Шанс, что у подсудимого была бы именно такая группа крови, если бы он был невиновен -- всего 1%; значит, с вероятностью 99% он виновен, предлагаю осудить”. В чём не прав прокурор? Укажите, какие вероятности он оценил, а какие должен был оценить.


##### Решение

Пусть BG = { у человека редкая группа крови }, а G = { человек виновен }.
Тогда:
<br>
$ P(BG) = 0.01$ 
<br>
$ P(BG | G) = 1 $
<br>
Прокурор посчитал вероятность $P(BG | \bar{G}) $ и принял ее равной 0.01 следуя предположению, что $ P(BG | \bar{G}) = P(A) $, что неверно.
<br>
В данном случае следовало бы говорить о вероятности $P(G | BG) $.

**Задача 2:**

Затем выходит адвокат, объясняет ошибки прокурора и берёт слово сам: “В городе живёт миллион человек. Это значит, что примерно у 10000 из них такая группа крови. Значит, всё, что говорит нам эта кровь -- это то, что подсудимый совершил убийство с вероятностью 0.01%. Получается, что это свидетельство не просто не является строгим доказательством, а и вовсе даёт нам ничтожно малую вероятность того, что мой подзащитный виновен, так что предлагаю исключить его из материалов дела”. В чём не прав адвокат? Какие вероятности он оценил, а какие должен был оценить?

##### Решение

Адвокат оценил вероятность того, что один из обладателей редкой группы крови является убийцей, т.е. $P(G|BG) = \frac{1}{10000} = 0,0001$. Вероятность встретить убийцу среди всех обладателей данной группы крови не позволяет сделать вывод о виновности/невиновности подозреваемого.
<br>
Адвокат должен посчитать вероятность $ P(\bar{G} | BG) $.

**Задача 3:**

Вернёмся к тесту на страшную болезнь из первой лекции. Напомню, что по предположениям задачи болеет 1% населения, а тест имеет вероятность ошибки 5% в обе стороны. Мы видели, что в результате человек с положительным тестом имеет апостериорную вероятность оказаться больным всего лишь около 16%.
<br>
Врачи решили, что терпеть это никак нельзя; они готовы провести срочное исследование и улучшить качество теста. Но сил и денег хватит только для того, чтобы сократить один тип ошибок, т.е. уменьшить или только число false positives (когда тест выдаёт положительный результат у здорового человека), или только число false negatives (когда тест выдаёт отрицательный результат у больного).
<br>
Помогите медикам: постройте зависимости апостериорной вероятности болезни после положительного и отрицательного теста от обоих видов ошибок (например, в виде графиков), сделайте выводы и дайте рекомендации врачам -- на чём лучше сосредоточиться?


Примем P = { тест положительный }, I = { человек заражен }. Тогда:
<br>
$P(P | \bar{I}) = x $ - false positive
<br>
$ P(\bar{P} | I) = y $ - false negative
<br>
По формуле Байеса:
<br>
$ P(I|P) = \frac{P(P|I) * P(I)}{P(P | I) * P(I) + P(P | \bar{I}) * P(\bar{I})} = \frac{0.01 * (1 - y)}{0.01 * (1 - y) + 0.99 * x} = p(x, y)$
<br>
$ P (I | \bar{P}) = \frac{P(\bar{P} | I) * P(I)}{P(\bar{P}, I)} = \frac{P(\bar{P} | I) * P(I)}{P(\bar{P} | I) * P(I) + P(\bar{P} | \bar{I}) * p(\bar{I})} = \frac{0,01*y}{0,01*y + 0,99*(1-x)} = q(x, y) $


```python
def prob_p(x, y):
    return 0.01 * (1 - y) / (0.01 * (1 - y) + 0.99 * x)

def prob_q(x, y):
    return 0.01 * y / ((0.01 * y) + 0.99 * (1 - x))
```

```python
fig, axs = plt.subplots(2, 2, figsize=(20, 10))

x_start = 0.05
y_start = 0.05


# False positives
x = np.arange(0, 0.051, 0.0001)
p = prob_p(x, y_start)
q = prob_q(x, y_start)
axs[0][0].plot(x, p)
axs[0][0].set_xlim(0, 0.05)
axs[0][0].set_ylim(0, 1)
axs[0][0].set_title(r"$P(I|P)$")
axs[0][0].set_xlabel("FP")
axs[0][0].set_ylabel("Prob")
axs[0][1].plot(x, q)
axs[0][1].set_xlim(0, 0.05)
axs[0][1].set_ylim(0, 0.001)
axs[0][1].set_title(r"$P(I|\bar{P})$")
axs[0][1].set_xlabel("FP")
axs[0][1].set_ylabel("Prob")


# False negatives
y = np.arange(0, 0.051, 0.0001)
p = prob_p(x_start, y)
q = prob_q(x_start, y)
axs[1][0].plot(y, p)
axs[1][0].set_xlim(0, 0.05)
axs[1][0].set_ylim(0, 0.2)
axs[1][0].set_title(r"$P(I|P)$")
axs[1][0].set_xlabel("FN")
axs[1][0].set_ylabel("Prob")
axs[1][1].plot(y, q)
axs[1][1].set_xlim(0, 0.05)
axs[1][1].set_ylim(0, 0.001)
axs[1][1].set_title(r"$P(I|\bar{P})$")
axs[1][1].set_xlabel("FN")
axs[1][1].set_ylabel("Prob")
```

**Вывод:**
<br>
1. Точность теста растет при уменьшении вероятности **false positive** ошибки.
2. Вероятность наличия болезни при отрицательном тесте **незначительно** зависит от вероятностей false positive и false negative ошибок.

**Пункт А:**

Используя линейную регрессию, обучите модель с экспоненциальным ростом числа заболевших: y ~ exp(линейная функция от x), где x — номер текущего дня.
<br>
Исходя из предположения экспоненциального роста:

**Пункт B:**
Найдите апостериорное распределение параметров этой модели для достаточно широкого априорного распределения. Требующееся для этого значение дисперсии шума в данных оцените, исходя из вашей же максимальной апостериорной модели.



Оценим дисперсию шума в данных по формуле: $$ \sigma^2 = \frac{1}{N}\sum_{n = 1}^N (y_n - w^T\Phi(x_n))^2 $$

```python
sigma_noise = np.mean((np.log(train_X["total_cases"]) - prediction) ** 2)
```

```python
def bayesian_update(mu, sigma, x, y, sigma_noise):
    sigma_n = np.linalg.inv(np.linalg.inv(sigma) + (1 / (sigma_noise ** 2)) * np.matmul(np.transpose(x), x) )
    mu_n = np.matmul(sigma_n, np.matmul(np.linalg.inv(sigma), np.transpose(mu)) + (1 / (sigma_noise ** 2)) * np.matmul(np.transpose(x), y))
    return mu_n, sigma_n
```

Параметры апостериорного распределения весов модели:
$$ \Sigma_N = (\Sigma_0^{-1} + \frac{1}{\sigma^2}X^TX)^{-1} $$
$$ \mu_N = \Sigma_N(\Sigma_0^{-1}\mu_0 + \frac{1}{\sigma^2}X^Ty) $$

```python
X = np.column_stack([np.ones(train_X.shape[0]), train_X["day"]])

mu, sigma = bayesian_update(np.zeros(2), 10**3 * np.eye(2), X, np.log(train_X["total_cases"]).values, sigma_noise)
```

**Задача 3:**
Предсказания экспоненциальной модели наверняка получились грустными. Но это, конечно, чересчур пессимистично — экспоненциальный рост в природе никак не может продолжаться вечно. Кривая общего числа заболевших во время эпидемии в реальности имеет сигмоидальный вид: после начальной фазы экспоненциального роста неизбежно происходит насыщение. В качестве конкретной формы такой сигмоиды давайте возьмём форму функции распределения для гауссиана:
$$ \Phi(x) = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{x} e^{\frac{-t^2}{2}}dt $$


**Пункт А:**
<br>
Константу перед сигмоидой можно внести в интеграл и расписать переменную t:
<br>
$$ y = \int_{-\infty}^{x} e^{ax^2 + bx + c} $$
<br>
Т.к. предсказания дискретны можно считать логарифм разницы между текущим и предыдущим днем:
$$ \log(y_i - y_{i - 1}) = ax^2 + bx +c $$
В качестве "фичей" используется номер дня и квадрат номера дня.

```python
lin_reg = LinearRegression()
X = np.column_stack([train_X["day"], train_X["day"] ** 2])
Y = train_X["new_cases"].copy()
Y.clip(lower=1, inplace=True)
Y = np.log(Y).values
```

```python
lin_reg.fit(X, Y)
```

```python
prediction = lin_reg.predict(X)
plt.figure(figsize=(10, 5))
plt.plot(train_X.index, train_X["total_cases"], label="Ground truth")
plt.plot(train_X.index, np.exp(prediction).cumsum(), label="Predicted")
plt.legend()
plt.xlabel("DATE")
plt.ylabel("TOTAL CASES")
```

**Задача 4:**
<br>
Бонус: проведите такой же анализ для других стран (здесь придётся руками подобрать дни начала моделирования — коронавирус приходил в разные страны в разное время). Насколько разные параметры получаются? Можно ли разделить страны на кластеры (хотя бы чисто визуально) в зависимости от этих параметров?


```python
TRAIN_DAYS = 50
infection_first_day = data[data["total_cases"] >= 2].groupby("iso_code").agg({"date": "min"}).reset_index()
infection_first_day.drop(infection_first_day[infection_first_day["iso_code"] == "RUS"].index, inplace=True)
infection_first_day.columns = ["iso_code", "start_date"]
```

```python
infection_first_day.head(10)
```

```python
filtered_data = data[["iso_code", "date", "total_cases", "new_cases"]].merge(infection_first_day, on="iso_code", how="inner")
filtered_data = filtered_data[filtered_data["date"] >= filtered_data["start_date"]]
filtered_data["day"] = filtered_data.groupby("iso_code").agg({"date": "cumcount"})
filtered_data["new_cases"].fillna(1, inplace=True)
filtered_data["new_cases"] = filtered_data["new_cases"].apply(lambda x: max(x, 1))
filtered_data = filtered_data.set_index("date")
filtered_data.head()
```

```python
mean, variance, models = {}, {}, {}
for iso in filtered_data["iso_code"].unique():
    current_data = filtered_data[filtered_data["iso_code"] == iso].iloc[:TRAIN_DAYS].reset_index(drop=True)
    
    lin_reg = LinearRegression()
    X = np.column_stack([current_data["day"], current_data["day"] ** 2])
    Y = current_data["new_cases"].copy()
    Y.clip(lower=1, inplace=True)
    Y = np.log(Y).values
    
    lin_reg.fit(X, Y)
    prediction = lin_reg.predict(X)
    
    sigma_noise = np.mean((Y - prediction) ** 2)
    x_matrix = np.column_stack([np.ones_like(train_X["day"]), X])
    mu, sigma = bayesian_update(np.zeros(3), np.eye(3), x_matrix, Y, sigma_noise)
    
    mean[iso] = mu
    variance[iso] = sigma
    models[iso] = lin_reg
```

```python
W = pd.DataFrame(mean).T
fig, axs = plt.subplots(1, 3, figsize=(10, 4))
for i in range(3):
    axs[i].hist(W[i], bins=15)
    axs[i].set_title(f"W[{i}] parameter distribution")
```

```python
fig, axs = plt.subplots(1, 3, figsize=(10, 4))
axs[0].scatter(W[0], W[1])
axs[0].set_title("W_0 and W_1 distribution")
axs[1].scatter(W[1], W[2])
axs[1].set_title("W_1 and W_2 distribution")
axs[2].scatter(W[0], W[2])
axs[2].set_title("W_0 and W_2 distribution")
```

Четко оформленных кластеров не неблюдается, только небольшое количество "выбросов".
